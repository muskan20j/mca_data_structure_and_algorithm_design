```python
def evaluate_postfix(expression):
    stack =[]

    for char in expression:
        if char.isdigit():           #operand
            stack.append(int(char))
        else:                        #operator
            b = stack.pop()
            a = stack.pop()     

            if char=='+':
               result = a+b
            elif char=='-':
               result = a-b
            elif char=='*':
               result = a*b
            elif char=='/':
               result=a/b

            stack.append(result)

    return stack.pop()

#example
exp = "23*54*+9-"
print(evaluate_postfix(exp))
              
```
